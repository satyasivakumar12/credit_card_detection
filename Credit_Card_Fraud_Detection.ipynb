**Importing Dependencies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Importing CSV file**

```python
Credit_Card_Data = pd.read_csv("/content/creditcard.csv")
```

**UNBALANCED DATA**

*   separating legit and fraud data
*   o--> legit data
*   1--> fraud data

```python
legit = Credit_Card_Data[Credit_Card_Data.Class == 0]
```

```python
fraud = Credit_Card_Data[Credit_Card_Data.Class == 1]
```

```python
legit.shape
```

```python
fraud.shape
```

**Statistical Measures Of The Data**

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

**Compare The Values For Both Transaction**

```python
Credit_Card_Data.groupby('Class').mean()
```

**On comparing both we can conclude that mean is devitating more for fraud transactions**

1.   so, we can use undersampling for legit transactions
2.   new dataset is also created







**Undersampling**

```python
legit_sample = legit.sample(n=492)
```

```python
legit_sample.shape
```

```python
new_dataset = pd.concat([legit_sample,fraud],axis = 0)
```

```python
print(new_dataset)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

**Splitting the data into features and targets**

```python
X = new_dataset.drop(columns='Class',axis = 1)
```

```python
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

**Splitting into training and testing data**

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

**Model training**

**Logistic Regression**

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression model with training data
model.fit(X_train,Y_train)
```

**Model Evaluation**

```python
# Accuracy score
# Accuracy on training data
X_train_prediction = model.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("training_data_accuracy :", training_data_accuracy)
```

```python
# Accuracy score
# Accuracy on testing data
X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("test_data_accuracy :", test_data_accuracy)
```

```python
test_data_accuracy_percentage = test_data_accuracy*100
```

```python
print(test_data_accuracy_percentage)
```
